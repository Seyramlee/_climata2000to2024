```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("bhadramohit/climate-change-dataset")

print("Path to dataset files:", path)
```

```python
print(path)
```

```python
import pandas as pd
df = pd.read_csv(path + '/climate_change_dataset.csv')
df.head()
```

```python
df.groupby('Country')['CO2 Emissions (Tons/Capita)'].mean().sort_values(ascending=False).head(10)
```

```python
df.groupby('Country')['Extreme Weather Events'].mean().sort_values(ascending=False).head(10)
```

```python
df.groupby('Country')['Renewable Energy (%)'].mean().sort_values(ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt

df.groupby('Country')['Renewable Energy (%)'].mean().sort_values(ascending=False).head(10).plot(kind='barh')
plt.title('Top 10 Countries by Renewable Energy %')
plt.xlabel('Renewable Energy %')
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df['Renewable Energy (%)'], df['CO2 Emissions (Tons/Capita)'])
plt.title('Renewable Energy vs CO2 Emissions')
plt.xlabel('Renewable Energy (%)')
plt.ylabel('CO2 Emissions (Tons/Capita)')
plt.show()
```
